```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,200)
y=np.sin(x)

fig,ax=plt.subplots()
#fig对应figure,为最外层的区域
#ax对应axes,为第二层区域
#subplots为布局管理器，管理图形对应的格式
##如果输入plt.subplots(1,4,figfize(12,8)),为生成1行4列，4个1200像素*800像素的图形

ax.plot(x,y,'b-')
ax.plot(x,0*x,'k--')
plt.show()
```

```python
fig,axes=plt.subplots(3,3,sharex=True,sharey=True)
axes[0,0].plot(x,y)
axes[0,2].scatter(x,y)
plt.show()
```

```python
fig,axes=plt.subplots(2,2,subplot_kw=dict(polar=True))
axes[0,0].plot(x,y)
axes[1,1].scatter(x,y)
fig.text(.45,1,'polar axes')
plt.show()
```

```python
#修改冒泡排序函数：
#1定义为一个函数，跟输入数据分开
#2增加默认参数，默认是顺序，也可以是逆序
#3增加默认参数是否显示冒泡的过程，如果要显示，应该每一轮冒泡后指出哪个数位置是如何变化的
nums=[4,6,8,2,3,8]
def bubble1(nums):
    for i in range(len(nums)):
        for j in range(len(nums)-i-1):#注意此处的range(len(nums)-i-1)
            if nums[j]>nums[j+1]:
                nums[j],nums[j+1]=nums[j+1],nums[j]
                print(nums)
                print(nums[j],r'与',nums[j+1],r'互换位置')
print(nums)
bubble1(nums)
print(nums)
```
